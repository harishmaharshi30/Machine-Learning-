 ## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22, 33, 44]

# 2. Setup plot
fix, ax = plt.subplots(figsize = (5, 5))

# 3. Plot data
ax.plot(x, y)

# 4 custmize plot
ax.set(title= "simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5.save and show (you save the whole figure)
fix.savefig("samapl-plot.png") 
```

## Two option for subplots

```python
# Subplot option : 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5));

# Plot to each diffrent axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
                                             
```

```python
fig,ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# Plot to each diffrent index
ax[0,0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(100));
```

### Which one should you use? (pyplot vs matplotlib OO method?)
.When plotting something quickly, okey to use the pyplot mehtod
.When Plottting more advanced ,use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[ heart_disease['age'] > 50]
over_50
```

```python
# Pyplot Method
over_50.plot( x= 'age',
             y = 'chol',
             c = 'target',
            kind = 'scatter');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
ax.set_xlim(45, 100)
```

```python
## Object Oriented method from scratch
fig,ax = plt.subplots(figsize =  (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax.set(title = 'Heart Disease and Cholesterol Leverls', 
       xlabel = 'Age',
       ylabel = 'Cholesterol')

# Add Legend
ax.legend( *scatter.legend_elements(), title = "Target");

# Add Horizontal Line
ax.axhline(over_50['chol'].mean(),
           linestyle = 'dashed');
```

```python
# Subplot of chol,age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 6))
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'])
# customize ax0
ax0.set(title = 'Heard Disease and Cholesterol Levels',
        xlabel = 'Age',
        ylabel = 'Cholesterol')
# Add Legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add the horizontal line t0 ax0
ax0.axhline( y = over_50['chol'].mean(),
            linestyle = '--');

ax1.scatter(x = over_50['age'],
            y = over_50['thalach'],
            c = over_50['target'])

# Customize ax1
ax1.set(title = "Heart Disease and Max Heart Rate",
       xlabel = "Age",
       ylabel = 'Thalach')

# Add legend to ax1
ax1.legend(* scatter.legend_elements(),
           title = 'Target', loc = 'upper right')

# Add title to the entire graph
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold')
```

```python
# set the style
plt.style.use('seaborn-v0_8-whitegrid')

# OO Method from the scratch
fix, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                    cmap = 'winter')

# Customize the plot
ax.set(title = 'Heart Disease and Cholestrol Leverls',
       xlabel = 'Age',
       ylabel = 'Cholestrol')

# add the horizontal line useing cholestrol mean()
ax.legend(*scatter.legend_elements(), title = 'Target');
       
```

```python
# Customizing the y and x axis limitations

# set style for plot
plt.style.use('seaborn-v0_8-whitegrid')


# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex = True)
# Add the super title to the plot
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold')

# Add Data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap = 'winter')

plt.style.use('seaborn-v0_8-whitegrid')
# Customize ax0
ax0.set(title = 'Heart Disease adn Cholesterol Levels',
        xlabel = 'age',
        ylabel = 'chol')
# chagne the  x-axis limits
ax0.set_xlim([50, 80])

# Add legend to ax0
ax0.legend(* scatter.legend_elements(), title = 'Target', loc= 'upper left')

# Add the Horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle = '--')

# Add data to ax1
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap = 'winter')
# Customize the ax1
ax1.set(title = 'Heart Deisease and Max Heart Rate',
        xlabel = 'age',
        ylabel = 'chol')

# customize the y lim
ax1.set_ylim([50, 200])

# Add a legend to ax1
ax1.legend(* scatter.legend_elements(), title = 'Target', loc = 'upper left' );

# Add Horizontal line using col.mean()
ax1.axhline(over_50['thalach'].mean(), linestyle = '--');

fig.savefig('hear_disease.png')
```
