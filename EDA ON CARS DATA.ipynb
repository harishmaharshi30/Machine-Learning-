# IMPORT THE LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes= True)
```

Loading the data into the dataframe

```python
df_cars = pd.read_csv(r"data.csv")
df_cars
```

```python
df_cars.head()
```

```python
df_cars.tail()
```

```python
df_cars.dtypes # To check the data types
```

# Dropping the Irrelevant columns

```python
df_cars_u = df_cars.drop(['Engine Fuel Type',"Market Category","Vehicle Style","Popularity","Number of Doors","Vehicle Size"],axis=1)
df_cars_u
```

# Renaming the columns

```python
df_cars_u = df_cars_u.rename(columns = {"Engine HP":"HP", "Engine Cylinders": "Cylinders",
                                      "Transmission Type": "Transmission","Driven_Wheels": "Drive Mode",
                                       "highway MPG": "MPG-H" , "city mpg": "MPG-C","MSRP" : "Price"})
df_cars_u.head()
```

```python
df_cars_u.shape
```

# Check the duplicated rows

```python
df_cars_u.count()
```

```python
duplicate_car = df_cars_u[df_cars_u.duplicated()]
print("Number of duplicate rows: ", duplicate_car.shape)
```

```python
duplicate_car
```

```python
df_cars_u = df_cars_u.drop_duplicates()
df_cars_u
```

```python
df_cars_u.count()
```

# Dropping the missing values

```python
df_cars_u.isnull().sum()
```

```python
df_cars_u = df_cars_u.dropna()
df_cars_u.count()
```

# Detecting the outliers

```python
# Output variable outliers

sns.boxplot(x = df_cars_u["Price"],orient="h")
```

```python
sns.boxplot(x = df_cars_u["HP"])
```

```python
sns.boxplot(x = df_cars_u["Cylinders"])
```

```python
Q1 = df_cars_u.quantile(0.25)
Q3 = df_cars_u.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5 * IQR
```

```python
price_upper = np.where(df_cars_u >= upper)[0]
price_lower = np.where(df_cars_u <= lower)[0]
```
